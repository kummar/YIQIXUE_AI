### 不使用正则化看看模型表现

**模型的参数都是之前提到过的常规参数，除了正则化系数‘lambd’，由于Python里面lambda是一个很重要的命令，所以用‘lambd’替代**

```python
## 定义模型框架
def model(X, Y, learning_rate = 0.3, num_iterations = 30000, print_cost = True, lambd = 0, keep_prob = 1):
    
    # 保留每层的梯度
    grads = {}
    
    # 储存cost
    costs = []
    
    # 得到样本数
    m = X.shape[1]     
    
    # 定义一个三层网络（是写死的）
    layers_dims = [X.shape[0], 20, 3, 1]
    
    # 初始化模型参数
    parameters = initialize_parameters(layers_dims)
    
    # 梯度下降
    
    for i in range(0, num_iterations):
        
        # 向前传播: LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID.
        
        if keep_prob == 1:
            a3, cache = forward_propagation(X, parameters)
        elif keep_prob < 1:
            a3, cache = forward_propagation_with_dropout(X, parameters, keep_prob) # dropout,后续会对其进行定义
            
        # 损失函数
        if lambd == 0:
            cost = compute_cost(a3, Y)
        else:
            cost = compute_cost_with_regularization(a3, Y, parameters, lambd)
            
        # 反向传播部分
        assert(lambd==0 or keep_prob==1)   # 保证dropout和L2正则化只用到了其中一个‘
        
        if lambd == 0 and keep_prob == 1:
            grads = backward_propagation(X, Y, cache)
        elif lambd != 0:
            grads = backward_propagation_with_regularization(X, Y, cache, lambd)
        elif keep_prob < 1:
            grads = backward_propagation_with_dropout(X, Y, cache, keep_prob)
            
        # 更新参数
        parameters = update_parameters(parameters, grads, learning_rate)
        
        # 每迭代10000次输出一次日志
        if print_cost and i % 10000 == 0:
            print("Cost after iteration {}: {}".format(i, cost))
        if print_cost and i % 1000 == 0:
            costs.append(cost)
            
    # plot the cost
    plt.plot(costs)
    plt.ylabel('cost')
    plt.xlabel('iterations (x1,000)')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()
    
    return parameters
```

```python
# 不使用任何正则化
parameters = model(train_X, train_Y)
print ("On the training set:")
predictions_train = predict(train_X, train_Y, parameters)
print ("On the test set:")
predictions_test = predict(test_X, test_Y, parameters)
```

**可以看到测试集的准确率是低于训练集的，我们可以通过绘制决策边界看看是不是由于过拟合造成的**

```python
# 绘制决策边界
plt.title("Model without regularization")
axes = plt.gca()
axes.set_xlim([-0.75,0.40])
axes.set_ylim([-0.75,0.65])
plot_decision_boundary(lambda x: predict_dec(parameters, x.T), train_X, train_Y)
```

### L2 正则化

和常规机器学习一样，我们只需要下面这个：

$$J = -\frac{1}{m} \sum\limits_{i = 1}^{m} \large{(}\small  y^{(i)}\log\left(a^{[L](i)}\right) + (1-y^{(i)})\log\left(1- a^{[L](i)}\right) \large{)} \tag{1}$$

变成

$$J_{regularized} = \small \underbrace{-\frac{1}{m} \sum\limits_{i = 1}^{m} \large{(}\small y^{(i)}\log\left(a^{[L](i)}\right) + (1-y^{(i)})\log\left(1- a^{[L](i)}\right) \large{)} }_\text{cross-entropy cost} + \underbrace{\frac{1}{m} \frac{\lambda}{2} \sum\limits_l\sum\limits_k\sum\limits_j W_{k,j}^{[l]2} }_\text{L2 regularization cost} \tag{2}$$

我们要对每一层的参数都做处理 $W^{[1]}$, $W^{[2]}$ 和 $W^{[3]}$, 然后把他们加起来再相乘 $ \frac{1}{m} \frac{\lambda}{2} $.

```python
def compute_cost_with_regularization(A3, Y, parameters, lambd):
    
    # 样本数
    m = Y.shape[1]
    
    # 得到权重
    W1 = parameters["W1"]
    W2 = parameters["W2"]
    W3 = parameters["W3"]
    
    # 重新定义损失函数
    cross_entropy_cost = compute_cost(A3, Y)
    
    L2_regularization_cost = lambd * (np.sum(np.square(W1)) + np.sum(np.square(W2)) + np.sum(np.square(W3))) / (2.0 * m)
    
    cost = cross_entropy_cost + L2_regularization_cost
    
    return cost
```

### 修改反向传播部分

得到新的 dW1, dW2 和 dW3. 
($\frac{d}{dW} ( \frac{1}{2}\frac{\lambda}{m}  W^2) = \frac{\lambda}{m} W$).

```python
def backward_propagation_with_regularization(X, Y, cache, lambd):
    
    m = X.shape[1]
    
    # 这部分都是常规的反向传播步骤，只是加了正则化项部分而已
    (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3) = cache
    
    dZ3 = A3 - Y
    dW3 = 1./m * np.dot(dZ3, A2.T) + (lambd * W3) / m
    db3 = 1./m * np.sum(dZ3, axis=1, keepdims = True)
    
    dA2 = np.dot(W3.T, dZ3)
    dZ2 = np.multiply(dA2, np.int64(A2 > 0))
    dW2 = 1./m * np.dot(dZ2, A1.T) + (lambd * W2) / m
    db2 = 1./m * np.sum(dZ2, axis=1, keepdims = True)
    
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = np.multiply(dA1, np.int64(A1 > 0))
    dW1 = 1./m * np.dot(dZ1, X.T) + (lambd * W1) / m
    db1 = 1./m * np.sum(dZ1, axis=1, keepdims = True)
    
    gradients = {"dZ3": dZ3, "dW3": dW3, "db3": db3,"dA2": dA2,
                 "dZ2": dZ2, "dW2": dW2, "db2": db2, "dA1": dA1, 
                 "dZ1": dZ1, "dW1": dW1, "db1": db1}
    
    return gradients    
```

```python
# 测试函数
X_assess, Y_assess, cache = backward_propagation_with_regularization_test_case()

grads = backward_propagation_with_regularization(X_assess, Y_assess, cache, lambd = 0.7)
print ("dW1 = "+ str(grads["dW1"]))
print ("dW2 = "+ str(grads["dW2"]))
print ("dW3 = "+ str(grads["dW3"]))
```

### 那么，为什么用了L2模型就变好了？

- L2依赖于一个假设：具有小权重的模型比拥有大权重的模型简单。
- 因此，通过惩罚成本函数中权重的平方值，可以使得所有权重趋向较小的值。
- 对cost来说，拥有大的权重值太昂贵了。
- 这样使得模型更加平滑，随着输入的改变，输出的变化会变得慢下来。
- **权重衰减**

### Dropout

- 当你随机关闭一些神经元时，实际上修改了模型的结构。 
- 在每次迭代中，训练的模型都是不同的，是所有神经元的一个子集。 
- 这样，你的神经元对另一个特定神经元的激活因变得不那么敏感，因为那个神经元随时可能关闭。

```python
# Dropout不会出现在输出层

def forward_propagation_with_dropout(X, parameters, keep_prob = 0.5):
    
    # 得到之前的参数，和L2不同的是，我们不需要再对W做操作，只需要在后面对神经元做操作
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]
    W3 = parameters["W3"]
    b3 = parameters["b3"]
    
    # LINEAR -> RELU -> LINEAR -> RELU -> LINEAR -> SIGMOID
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    
    D1 = np.random.rand(A1.shape[0], A1.shape[1]) # 随机                    
    D1 = D1 < keep_prob                                         
    A1 = A1 * D1                            
    A1 = A1 / keep_prob  
```
