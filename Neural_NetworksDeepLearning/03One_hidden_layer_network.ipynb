```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import sklearn.datasets
import sklearn.linear_model
import pandas as pd

%matplotlib inline

np.random.seed(1) # set a seed so that the results are consistent
```

```python
# 手动生成一些数据用于训练

def load_planar_dataset():
    np.random.seed(1)
    m = 400 # number of examples
    N = int(m/2) # number of points per class
    D = 2 # dimensionality
    X = np.zeros((m,D)) # data matrix where each row is a single example
    Y = np.zeros((m,1), dtype='uint8') # labels vector (0 for red, 1 for blue)
    a = 4 # maximum ray of the flower

    for j in range(2):
        ix = range(N*j,N*(j+1))
        t = np.linspace(j*3.12,(j+1)*3.12,N) + np.random.randn(N)*0.2 # theta
        r = a*np.sin(4*t) + np.random.randn(N)*0.2 # radius
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        Y[ix] = j
        
    X = X.T
    Y = Y.T

    return X, Y
```

```python
# 得到数据并查看一下数据分布
X, Y = load_planar_dataset()


shape_X = X.shape
shape_Y = Y.shape
m = Y.shape[1] 

print ('The shape of X is: ' + str(shape_X))
print ('The shape of Y is: ' + str(shape_Y))
print ('I have m = %d training examples!' % (m))
```

```python
def plot_decision_boundary(model, X, y):
    # Set min and max values and give it some padding
    x_min, x_max = X[0, :].min() - 1, X[0, :].max() + 1
    y_min, y_max = X[1, :].min() - 1, X[1, :].max() + 1
    h = 0.01
    # Generate a grid of points with distance h between them
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    # Predict the function value for the whole grid
    Z = model(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    # Plot the contour and training examples
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral)
    plt.ylabel('x2')
    plt.xlabel('x1')
    plt.scatter(X[0, :], X[1, :], c=y, cmap=plt.cm.Spectral)
```

```python
# 然后我们看看用LR分类的结果如何

plot_decision_boundary(lambda x: clf.predict(x), X, Y)
plt.title("Logistic Regression")

# Print accuracy
LR_predictions = clf.predict(X.T)
print ('Accuracy of logistic regression: %d ' % float((np.dot(Y, LR_predictions) + np.dot(1 - Y,1 - LR_predictions)) / float(Y.size) * 100) +
       '% ' + "(percentage of correctly labelled datapoints)")
```

LR效果差的主要原因我们上面已经提到过了，就是因为这些样本点本身就是线性不可分的，下面我们尝试用含有一个隐藏层神经网络做一次分类看看效果。

我们先要根据下图设计出网络的结构：
<img src="classification_kiank.png" style="width:600px;height:300px;">

看起来也不是很复杂，我们只需要遵循之前课程提到的计算逻辑，一步步实现下面几个公式就可以了:
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1] (i)}\tag{1}$$ 
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2] (i)}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$

有了上面的整体计算流程，我们再加上下面的损失函数定义，很快就可以实现一个神经网络了
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

**我们再说一下神经网络的训练流程**
1. 设计好整个网络的架构
2. 初始化所有参数
3. 循环过程:
    - 向前传播
    - 计算损失函数
    - 反向传播求导
    - 用梯度下降更新参数

```python
# 设计网络的size
#  n_x: 输入的SIZE
#  n_h: 隐藏层的SIZE
#  n_y: 输出的SIZE

def layer_sizes(X, Y):

    
    n_x = X.shape[0] # size of input layer
    n_h = 4 # 上图中隐藏层就四个单元
    n_y = Y.shape[0] # size of output layer

    return (n_x, n_h, n_y)
```

```python
# 初始化参数
# 这个步骤最重要的地方就是SIZE，一定要设定好SIZE，不然就会报错
# 我们用Numpy来随机生成参数作为初始化参数


def initialize_parameters(n_x, n_h, n_y):

    # 设定种子，方便回溯
    np.random.seed(2) # we set up a seed so that your output matches ours although the initialization is random.
    
    # 两层的W和b
    W1 = np.random.randn(n_h, n_x) * 0.01
    b1 = np.zeros(shape = (n_h, 1))
    W2 = np.random.randn(n_y, n_h) * 0.01
    b2 = np.zeros(shape = (n_y, 1))

    # 确认SIZE是否正确
    assert (W1.shape == (n_h, n_x))
    assert (b1.shape == (n_h, 1))
    assert (W2.shape == (n_y, n_h))
    assert (b2.shape == (n_y, 1))
    
    # 以字典的形式，得到相应的参数
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

```python
# 向前传播
# 需要定义一个sigmoid函数作为最后的输出，函数的实现方式上一个notebook已经有说明
# 用Numpy调用tanh做激活函数

def forward_propagation(X, parameters):

    # 注意，这部分的参数都是之前初始化而来的参数
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']

    # 计算A1，A2的值
    Z1 = np.dot(W1, X) + b1
    A1 = np.tanh(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = sigmoid(Z2)

    
    assert(A2.shape == (1, X.shape[1]))
    
    # 同样得到一个计算后的值，以字典的形式给出
    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}
    
    return A2, cache
```

```python
# sigmoid函数
def sigmoid(x):
    
    s = 1.0 /(1+np.exp(-x))
    
    return s
```

```python
# 接着就是计算损失函数了
# 我们用到的是上述公式的交叉熵损失函数
# 同样，用Numpy实现中间的计算过程

def compute_cost(A2, Y, parameters):
    
    m = Y.shape[1] # number of example
    
    # 计算交叉熵损失函数，这部分有用到之前提到的Numpy的broadcasting
    cost = np.sum(np.multiply(np.log(A2), Y) + np.multiply(np.log(1-A2), (1-Y))) / (-1 * m)

    # 确定我们得到的结果SIZE是合适的
    cost = np.squeeze(cost)    
    assert(isinstance(cost, float))
    
    return cost
```

**然后我们需要计算反向传播**

按照下面公式计算即可

```python
# 反向传播

def backward_propagation(parameters, cache, X, Y):

    m = X.shape[1]
    
    
    W1 = parameters['W1']
    W2 = parameters['W2']

    A1 = cache['A1']
    A2 = cache['A2']

    
    # 计算dW1, db1, dW2, db2. 
    dZ2 = A2 - Y
    dW2 = (1.0 / m) * np.dot(dZ2, A1.T)
    db2 = (1.0 / m) * np.sum(dZ2, keepdims=True, axis=1)
    dZ1 = np.multiply(np.dot(W2.T, dZ2), 1 - np.power(A1, 2))
    dW1 = (1.0 / m) * np.dot(dZ1, X.T)
    db1 = (1.0 / m) * np.sum(dZ1, axis=1, keepdims=True)
    ### END CODE HERE ###
    
    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}
    
    return grads
```

**接下来就要开始更新参数了,有了上面几个部分，这个函数是很好写的，只需定一个学习率每次更新就行** 

```python
def update_parameters(parameters, grads, learning_rate=1.2):

    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']

    dW1 = grads['dW1']
    db1 = grads['db1']
    dW2 = grads['dW2']
    db2 = grads['db2']
    
    # 通过定好学习率更新参数
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2
    
    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}
    
    return parameters
```

**通过上面所有的函数，我们已经讲神经网络框架中所有的计算流程都定义好了，把这些定义好的函数组合在一起就是神经网络的模型了**

```python
# nn_model

def nn_model(X, Y, n_h, num_iterations=10000, print_cost=False):

    
    np.random.seed(3)
    n_x = layer_sizes(X, Y)[0]
    n_y = layer_sizes(X, Y)[2]
    
    # Initialize parameters, then retrieve W1, b1, W2, b2. Inputs: "n_x, n_h, n_y". Outputs = "W1, b1, W2, b2, parameters".
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']

    
    # Loop (gradient descent)
    for i in range(0, num_iterations):
         
        # Forward propagation. Inputs: "X, parameters". Outputs: "A2, cache".
        A2, cache = forward_propagation(X, parameters)
        
        # Cost function. Inputs: "A2, Y, parameters". Outputs: "cost".
        cost = compute_cost(A2, Y, parameters)
 
        # Backpropagation. Inputs: "parameters, cache, X, Y". Outputs: "grads".
        grads = backward_propagation(parameters, cache, X, Y)
 
        # Gradient descent parameter update. Inputs: "parameters, grads". Outputs: "parameters".
        parameters = update_parameters(parameters, grads)
        
        
        # Print the cost every 1000 iterations
        if print_cost and i % 1000 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
            

    return parameters
```

注意一下sigmoid的预测过程！
predictions = $y_{prediction} = \mathbb 1 \text{{activation > 0.5}} = \begin{cases}
      1 & \text{if}\ activation > 0.5 \\
      0 & \text{otherwise}
    \end{cases}$  

```python
# predict

def predict(parameters, X):
    
    A2, cache = forward_propagation(X, parameters)
    predictions = (A2 > 0.5)
    
    return predictions
```
